```python
movies = ['JL', 'Batman','Avengers','Ironman','Fast&Furious']
percentage=[90,80,77,99,100]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Scatter plot

```python
plt.scatter(movies,percentage)
```

# Line plot

```python
plt.plot(movies,percentage,linestyle='solid')
```

```python
plt.pie(percentage , labels=movies)
```

```python
plt.bar(percentage,movies)
```

```python
plt.bar(movies,percentage)
```
