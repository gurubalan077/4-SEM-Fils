# Working with inbuild datasets

```python
sns.get_dataset_names()
```

```python
tip1 = sns.load_dataset('tips')
```

```python
tip1
```

```python
tip1.tail()
```

```python
tip1.head()
```

# rels plot

```python
sns.relplot(x='total_bill',y = 'tip',data=tip1,col='time',hue='day',
           style='day',kind='scatter')
```

```python
sns.relplot(x='total_bill',y = 'tip',data=tip1,col='time',hue='day',
           style='day',kind='line') # just you know
```

```python
sns.catplot(x='day' , y = 'total_bill', kind = 'box' , data = tip1)
```

# Regression plot

```python
sns.regplot(x='total_bill',y='tip',data=tip1)
```

```python
sns.regplot(x='tip',y='size',data=tip1) # just for you
```

```python
sns.regplot(x='total_bill',y='size',data=tip1) # just for you
```
