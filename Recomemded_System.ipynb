```python
moviemat = df.pivot_table(index = 'user_id' , columns = 'title' , values='rating')
moviemat.head()
```

```python
rating.sort_values('num of rating' , ascending = False).head()
```

```python
starwar_user_ratings = moviemat['Star Wars (1977)']
liarliar_user_ratings = moviemat['Liar Liar (1997)']
starwar_user_ratings.head(50)
```

```python
liarliar_user_ratings.head(50)
```

```python
similar_to_starwars = moviemat.corrwith(starwar_user_ratings)
similar_to_liarliar = moviemat.corrwith(liarliar_user_ratings)
```

```python
corr_starwar = pd.DataFrame(similar_to_starwars , columns =['correlation'] )
corr_starwar.dropna(inplace=True)
corr_starwar.head()
```

```python
corr_starwar.sort_values('correlation',ascending = False).head(50)
```

```python
corr_starwar = corr_starwar.join(rating['num of rating'])
```

```python
corr_starwar.head()
```

```python
corr_starwar[corr_starwar['num of rating'] >100].sort_values('correlation' , ascending = False).head()
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar , columns =['correlation'] )
corr_liarliar.dropna(inplace=True)
corr_liarliar = corr_liarliar.join(rating['num of rating'])
corr_liarliar[corr_liarliar['num of rating'] >100].sort_values('correlation' , ascending = False).head()
```
