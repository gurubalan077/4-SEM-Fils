```python
import numpy as np
import pandas as pd
```

```python
stand_scaler = pd.DataFrame({'x1':np.random.normal(0,2,100),
                            'x2':np.random.normal(3,5,100),
                            'x3':np.random.normal(-2,2,100)})
```

```python
stand_scaler
```

```python
stand_scaler.plot.kde()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
stand_sca = StandardScaler()
```

```python
stand_scaler1 = stand_sca.fit_transform(stand_scaler)
```

```python
stand_scaler1
```

```python
stand_scaler2 = pd.DataFrame(stand_scaler1,columns=['x1','x2','x3'])
```

```python
stand_scaler2
```

```python
stand_scaler2.plot.kde()
```

# Outlayer Scaling

```python
outlier_df = pd.DataFrame({
    'x1':np.concatenate([np.random.normal(20,1,1000),np.random.normal(1,1,25)]),
    'x2':np.concatenate([np.random.normal(30,1,1000),np.random.normal(50,1,25)]),
    })
```

```python
outlier_df
```

```python
outlier_df.plot.kde()
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
rs = RobustScaler()
```

```python
outlier_df1 = rs.fit_transform(outlier_df)
```

```python
outlier_df2 = pd.DataFrame(outlier_df1,columns=['x1','x2'])
```

```python
outlier_df2.plot.kde()
```
